```python
# Importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
# Lendo o arquivo do excel e jogando em um dataframe
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
# Mostrando as primeiras linhas do dataframe
df.head()
```

```python
# Quantidade de linhas e colunas do dataframe
df.shape
```

```python
# Verificando os tipos de dados
df.dtypes
```

```python
# Qual foi a receita total
df["Valor Venda"].sum()
```

```python
# Custo total da venda individual
df["Custo"] = df["Custo Unitário"].mul(df["Quantidade"])
df.head(1)
```

```python
# Custo total de todos os produtos vendidos - método round arredonda valores par n casas decimais
round(df["Custo"].sum(), 2)
```

```python
# Calculando e armazenando o valor do lucro para cada venda
df["Lucro"] = df["Valor Venda"] - df["Custo"]
df.head()
```

```python
# Calculando o lucro total
round(df["Lucro"].sum(), 2)import matplotlib.pyplot as plt
```

```python
# Criando coluna com total de dias gasto para enviar o produto
df["Tempo_Envio"] =  df['Data Envio'] - df['Data Venda']
df.head(1)
```

```python
# Verificando os tipod de dados das colunas
df.dtypes
```

```python
# Pegando a diferença de dias de forma que o campo criado seja numérico
df["Tempo_Envio"] = (df['Data Envio'] - df['Data Venda']).dt.days
```

```python
# Verificando o tipo de dados da coluna Tempo_Envio
df["Tempo_Envio"].dtypes
```

```python
# Média de tempo de envio por marca
round(df.groupby("Marca")["Tempo_Envio"].mean(), 2)
```

```python
# Verificar se existe campos sem preencher
df.isnull().sum()
```

```python
# Agrupar por ano e marca
df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum()
```

```python
# Utilizando o pandas para formatar números em ponto flutuante

pd.options.display.float_format = '{:20,.2f}'.format
```

```python
# Resetando o index e trazendo o resultado em forma de dataframe coluna
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum().reset_index()
lucro_ano
```

```python
# Filtrar por Total de produtos vendidos e ordenar por quantidade
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

```python
# Mostrar gráfico de Total de produtos vendidos e ordenar por quantidade
df.groupby("Produto")["Quantidade"].sum().plot.barh(title="Total de Produtos Vendidos")
plt.xlabel("Quantidade")
plt.ylabel("Produto")
```

```python
# Mostrar gráfico de Lucro por ano
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum().plot.bar(title="Receita por Ano")
plt.xlabel("Receita")
plt.ylabel("Ano");
```

```python
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum()
```

```python
# Criando um dataframe apenas com dados de 2009
df_2009 = df[df["Data Venda"].dt.year == 2009]
```

```python
# Mostrando os dados do dataframe df_2009
df_2009.head(1)
```

```python
# Mostrar no gráfico o lucro por mês
df_2009.groupby(df_2009["Data Venda"].dt.month)["Lucro"].sum().plot(title="Lucro de 2009")
plt.xlabel("Mês")
plt.ylabel("Lucro")
```

```python
# Gráfico de Lucro por Marca
df_2009.groupby("Marca")["Lucro"].sum().plot.bar(title="Lucro X Marca")
plt.xlabel("Marca")
plt.ylabel("Lucro")
plt.xticks(rotation='horizontal');
```

```python
# Gráfico de Lucro por classe
df_2009.groupby("Classe")["Lucro"].sum().plot.bar(title="Lucro X Classe")
plt.xlabel("Classe")
plt.ylabel("Lucro")
plt.xticks(rotation='horizontal');
```
